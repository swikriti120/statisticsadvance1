1.What is a random variable in probability theory?
- A random variable in probability theory is a variable that represents the outcomes of a random experiment. It assigns numerical values to each possible outcome of the experiment.

3.What is the difference between discrete and continuous distributions?
- Discrete distributions:- A discrete distribution describes a random variable that can take on a finite or countable number of distinct values.                 
Continuous distributions:- A continuous distribution describes a random variable that can take on an infinite number of values within a range or interval.
4.What are probability distribution functions(PDF)?
- Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a random variable. The concept of a PDF differs slightly depending on whether the random variable is discret.
5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- Cumulative distiribution functions(CDF) Describes the probability that a random variable will take on a value less than or equal to a certain value.
It is cumulative, meaning it aggregates probabilities up to a given point.  
Probability distribution function (PDF) Describes the likelihood of a random variable taking on specific values (discrete case) or within a very small range of values (continuous case).
6.What is a discrete uniform distribution?
- A discrete uniform distribution is a probability distribution in which a finite number of outcomes are equally likely. Every value in the range of the random variable has the same probability of occurring.
7.What are the key properties of a Bernoulli distribution?
- The Bernoulli distribution is a simple and fundamental probability distribution that describes a random experiment with only two possible outcomes: success (1) or failure (0).

8.What is the binomial distribution, and how is it used in probability?
- The binomial distribution is a probability distribution that models the number of successes in a fixed number of independent trials, where each trial has the same probability of success. It is widely used in probability and statistics for modeling binary outcomes (success/failure) across repeated experiments.

9.What is the Poisson distribution and where is it applied?
- The Poisson distribution is a probability distribution that models the number of events occurring in a fixed interval of time, space, or other measure, under the assumption that the events happen at a constant average rate and independently of each other.

10.What is a continuous uniform distribution?
- The continuous uniform distribution is a probability distribution in which all outcomes within a specified range are equally likely, and the probability density is constant across this range. It is one of the simplest continuous probability distributions.

11.What are the characteristics of a normal distribution?
- The normal distribution (also called the Gaussian distribution) is one of the most important and widely used probability distributions in statistics. It describes a continuous random variable where most of the observations cluster around the mean, and the probability decreases symmetrically as we move further from the mean.

12.What is the standard normal distribution, and why is it important?
- The standard normal distribution is a special case of the normal distribution with specific properties that make it extremely useful in statistical analysis. It is a normal distribution where the mean is 0 and the standard deviation is 1. This distribution is often denoted by Z.

13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The CLT states that if you take a sufficiently large number of independent random samples from any population with any shape of distribution (as long as the population has a finite mean and variance), the sampling distribution of the sample mean will approximate a normal distribution. This holds true regardless of the original distribution of the population.

14.How does the Central Limit Theorem relate to the normal distribution
- The Central Limit Theorem (CLT) and the normal distribution are closely related, and understanding their connection is fundamental to many statistical methods. This means that even if the population distribution is not normal, the sampling distribution of the sample mean will tend toward a normal distribution as the sample size increases.

15.What is the application of Z statistics in hypothesis testing?
- The Z-statistic (or Z-test) is a widely used statistical tool for hypothesis testing, particularly when the data are approximately normally distributed or when the sample size is large (n>30). It is used to determine whether a sample mean significantly differs from a hypothesized population mean.

16.How do you calculate a Z-score, and what does it represent?
- A Z-score represents the number of standard deviations a data point is from the mean of a distribution. It allows we to compare individual data points to a distribution and standardize values for comparison across different datasets.

The formula to calculate a Z-score is:                                          
    Z=(X-μ)/σ

Where:

Z: The Z-score
X: The data point (value you are analyzing)
μ: The population mean (average of the dataset)
σ: The population standard deviation (spread of the dataset)
Interpretation of a Z-Score

17.What are point estimates and interval estimates in statistics?
- A point estimate is a single value calculated from the sample data that serves as an estimate for a population parameter. It provides a "best guess" of the parameter's value
An interval estimate provides a range of values (an interval) within which the true population parameter is expected to lie, along with a specified level of confidence.

18.What is the significance of confidence intervals in statistical analysis?
- Confidence intervals (CIs) are crucial in statistical analysis because they provide a range of plausible values for a population parameter, allowing researchers to make informed decisions and account for uncertainty in their estimate.

19.What is the relationship between a Z-score and a confidence interval?
- The relationship between a Z-score and a confidence interval lies in how the Z-score determines the width of the confidence interval in a standard normal distribution. The Z-score specifies the number of standard deviations away from the mean required to capture a desired level of confidence. This relationship is central to constructing confidence intervals for population parameters.
Higher confidence levels require larger Z-scores, which increase the margin of error and result in wider confidence intervals.
Lower confidence levels use smaller Z-scores, leading to narrower confidence intervals.

20.How are Z-scores used to compare different distributions?
- Z-scores are used to compare different distributions by standardizing values from different datasets or distributions to a common scale. This standardization allows comparisons between values that may originally come from distributions with different means and standard deviations.

22.What is the concept of expected value in a probability distribution?
- The expected value of a probability distribution is a measure of the central tendency, representing the long-term average or mean value of a random variable if an experiment is repeated infinitely many times. It is a fundamental concept in probability and statistics.
